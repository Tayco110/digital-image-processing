```python
#imports
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import math
```

```python
#Import to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
#plot
def plotImages(images):
    _, ax = plt.subplots(1, len(images), figsize=(5 * len(images), 5))
    for i in range(len(images)):
            ax[i].axis('off')
            ax[i].imshow(images[i])
    plt.show()
```

# Resultado 1

```python
img = cv.imread("building.tif")
```

```python
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
Gx =  cv.filter2D(img, cv.CV_16S, sobel_x)
Gy =  cv.filter2D(img, cv.CV_16S, sobel_y)
result_1 = np.abs(Gx) + np.abs(Gy)
resultado_1 = [img, np.abs(Gx), np.abs(Gy), result_1]
```

```python
plotImages(resultado_1)
```

## Resultado 2

```python
alpha = np.arctan2(Gy,Gx)
resultado_2 = [img, alpha]
plotImages(resultado_2)
```

## Resultado 3

```python
s_kernel = np.ones((5,5),np.float32)/25
s_img = cv.filter2D(img,cv.CV_16S, s_kernel)
```

```python
sGx =  cv.filter2D(s_img, cv.CV_16S, sobel_x)
sGy =  cv.filter2D(s_img, cv.CV_16S, sobel_y)
result_3 = np.abs(sGx) + np.abs(sGy)
resultado_3 = [s_img, np.abs(sGx), np.abs(sGy), result_3]
```

```python
plotImages(resultado_3)
```

## Resultado 4

```python
sobel_xd = np.array([[-2,-1,0],[-1,0,1],[0,1,2]])
sobel_yd = np.array([[0,1,2],[-1,0,1],[-2,-1,0]])
```

```python
Gxd =  cv.filter2D(img, cv.CV_16S, sobel_xd)
Gyd =  cv.filter2D(img, cv.CV_16S, sobel_yd)
resultado_4 = [np.abs(Gxd), np.abs(Gyd)]
```

```python
plotImages(resultado_4)
```

## Resultado 5

```python
_,t_n = cv.threshold(result_1, 0.33*result_1.max(),255, cv.THRESH_BINARY)
_,t_s = cv.threshold(result_3, 0.33*result_3.max(),255, cv.THRESH_BINARY)
result_5 = [t_n, t_s]
```

```python
plotImages(result_5)
```
