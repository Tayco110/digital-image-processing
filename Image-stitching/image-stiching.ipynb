```python
#imports
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import cv2
```

```python
#setup
ic1 = cv2.cvtColor(np.array(cv2.imread(r'images\ic1.jpg')),cv2.COLOR_BGR2RGB)
ic2 = cv2.cvtColor(np.array(cv2.imread(r'images\ic2.jpg')),cv2.COLOR_BGR2RGB)
ic3 = cv2.cvtColor(np.array(cv2.imread(r'images\ic3.jpg')),cv2.COLOR_BGR2RGB)
ic4 = cv2.cvtColor(np.array(cv2.imread(r'images\ic4.jpg')),cv2.COLOR_BGR2RGB)
images = np.array([ic1,ic2,ic3,ic4])
```

```python
#unstitched images
_, ax = plt.subplots(1, len(images), figsize=(5 * len(images), 5))
for i in range(len(images)):
    ax[i].imshow(images[i])
plt.show()
```

```python
#stiching images
stitcher = cv2.Stitcher_create()
_, stitchedImage = stitcher.stitch(images)
#cv2.imwrite('stiched-IC.png',stitchedImage)
```

```python
#result
figure(figsize=(5 * len(images), 5))
plt.imshow(stitchedImage)
plt.show()
```

```python
#Cropping the result
uncroppedIC = np.array(stitchedImage)
croppedIC = uncroppedIC[300:2650,100:9300]
#cv2.imwrite('stiched-IC-cropped.png',croppedIC)
```

```python
#result
figure(figsize=(5 * len(images), 5))
plt.imshow(croppedIC)
plt.show()
```
