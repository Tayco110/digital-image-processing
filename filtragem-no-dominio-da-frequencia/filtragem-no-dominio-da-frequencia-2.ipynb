```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Import to ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
#carregando imagem base
img = cv.imread('lena.png',0)
plt.imshow(img, cmap='gray')
rows, col = img.shape
```

```python
#creating noise
x = np.arange(0,rows,15)
y = np.arange(0,col,15)
X, Y = np.meshgrid(x, y)
Z = np.sin(X)
```

```python
#ploting noise
fgr_1 = Axes3D(plt.figure())
fgr_1.plot_surface(X,Y,Z, color = 'k',alpha = 0.1)
plt.show()
```

```python
#noise image result
plt.contourf(X,Y,Z, cmap = 'gray')
plt.axis('scaled')
plt.show()
```

```python
#rotated noise result
noise = cv.imread('noise.png',0)
theta = -45
height, width= noise.shape
rotationMatrix = cv.getRotationMatrix2D((width/2, height/2), theta, 1.5)

rotated = cv.warpAffine(noise, rotationMatrix,(width, height))
newDimension = np.array([rows,col])
resizedNoise = cv.resize(rotated, newDimension)

plt.imshow(resizedNoise, cmap='gray')
```

```python
#aplicando o ruido
resultWithNoise = cv.bitwise_and(img, resizedNoise)
plt.imshow(resultWithNoise, cmap='gray')
```
